```python
import numpy as np
import tensorflow as tf

# Create the dataset
data = np.array([
    [0, 0, 1, 25, 70, 6.5, 10, 0],
    [0, 1, 1, 28, 65, 6.8, 5, 1],
    [1, 0, 1, 20, 80, 6.0, 15, 1],
    [1, 1, 0, 22, 75, 6.2, 12, 1]
])

# Split the dataset into features and labels
X = data[:, :-1]
y = data[:, -1]

# Define the neural network model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(4, input_dim=X.shape[1], activation='sigmoid'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

# Compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
model.fit(X, y, epochs=1000, verbose=1)

# Test the model
test_data = np.array([
    [0, 0, 1, 24, 72, 6.3, 8],
    [0, 1, 1, 27, 68, 6.7, 7],
    [1, 0, 1, 21, 78, 6.1, 14],
    [1, 1, 0, 23, 74, 6.4, 11]
])

predictions = model.predict(test_data)
rounded_predictions = np.round(predictions)

for i in range(len(test_data)):
    print(f"Input: {test_data[i]} | Predicted Output: {rounded_predictions[i][0]}")
```

```python
import matplotlib.pyplot as plt
history = model.fit(X, y, epochs=1000, verbose=1)
plt.plot(history.history['loss'])
plt.title('Training Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()
```

```python
x_values = np.arange(len(test_data))
plt.scatter(x_values, rounded_predictions, color='red', label='Predicted Output')
plt.scatter(x_values, test_data[:, -1], color='blue', label='Actual Output')
plt.xticks(x_values, ['Sample 1', 'Sample 2', 'Sample 3', 'Sample 4'])
plt.title('XOR Gate Predictions')
plt.xlabel('Samples')
plt.ylabel('Output')
plt.legend()
plt.show()
```
