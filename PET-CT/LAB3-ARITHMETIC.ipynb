# PET and CT Image Arithmetic Operations with OpenCV

## Introduction

This code demonstrates how to perform various arithmetic operations on PET and CT images using OpenCV and visualize the results using matplotlib. The code also includes automatic resizing of images to ensure they have the same dimensions before performing the operations.

## Code Structure

1. **Image Loading**: PET and CT images are loaded using OpenCV's `cv2.imread`. It checks if the images were loaded successfully.

2. **Image Resizing**: The code automatically resizes the CT image to match the dimensions of the PET image if they are different. This ensures that arithmetic operations can be performed.

3. **Arithmetic Operations**:
   - Addition: `cv2.add`
   - Subtraction: `cv2.subtract`
   - Multiplication: `cv2.multiply`
   - Division: `cv2.divide`
   - Bitwise AND: `cv2.bitwise_and`
   - Bitwise OR: `cv2.bitwise_or`
   - Bitwise XOR: `cv2.bitwise_xor`

4. **Visualization with Matplotlib**:
   - The original PET and CT images, as well as the results of the arithmetic operations, are displayed using matplotlib's `plt.imshow`.
   - Subplots are used to arrange the images in a 3x3 grid for better visualization.
   - The `plt.title` function sets titles for each subplot.
   - `plt.tight_layout()` ensures proper spacing between subplots.

5. **Error Handling**: Checks are in place to handle errors when loading images and when the images have mismatched sizes.
